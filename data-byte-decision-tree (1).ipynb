Creating decision tree 

Node class

```python
class node:
    def __init__(self,feature_index=None,threshold=None,left=None,right=None,info_gain=None,value=None):
        #for decision node 
        self.feature_index=feature_index
        self.threshold=threshold
        self.left=left
        self.right=right
        self.info_gain=info_gain
        
        #for leaf node
        self.value=value
```

 Tree class

```python
class decision_tree:
    def __init__(self,min_sample=5,max_depth=5):
        self.root=None
        self.min_sample=min_sample
        self.max_depth=max_depth
        
    def creating_tree(self,datasheet,curr_depth=0):
        x,y=datasheet[:,:-1],datasheet[:,-1]
        no_samples,no_features=np.shape(x)
        
        #splitting the tree untill conditions satisfied 
        
        if no_samples>=self.min_sample and curr_depth<=self.max_depth:
            
            best_split=self.get_best_split(datasheet,no_samples,no_features)
    
            if best_split['info_gain']>0:
                left_sub_tree=self.cretaing_tree(best_split['datasheet_left'],curr_depth+1)
                right_sub_tree=self.creating_tree(best_split['datasheet_right'],curr_depth+1)
                return node(best_split['feature_index'],best_split['threshold'],left_sub_tree,right_sub_tree,best_split['info_gain'])
            
        leaf_value=self.get_leaf_value(y)
        return node(value=leaf_value)
    
    def get_best_split(self,datasheet,no_samples,no_features):
        best_split={}
        max_info_gain=-float('inf')
        
        for feature_index in range(no_features):
            feature_value=datasheet[:,feature_index]
            possible_threshold=np.unique(feature_value)
            
            for threshold in possible_threshold:
                datasheet_left,datasheet_right=self.split(datasheet,feature_index,threshold)
                y,y_left,y_right=datasheet[:,-1],datasheet_left[:,-1],datasheet[:,-1]
                curr_info_gain=self.information_gain(y,y_left,y_right,"gini")
                
                if curr_info_gain>max_info_gain:
                    best_split['feature_index']=feature_index
                    best_split['threshold ']=threshold 
                    best_split['datasheet_left']=datasheet_left
                    best_split['datasheet_right']=datasheet_right
                    best_split['info_gain']=curr_info_gain
                    max_info_gain=curr_info_gain
                
        return best_split
    
    def split(self,datasheet,feature_index,threshold):
        
        datasheet_left=np.array([row for row in datasheet if row[feature_index]<=threshold])
        datasheet_right=np.array([row for row in datasheet if row[feature_index]>threshold])
        return datasheet_left,datasheet_right
    
    def information_gain(self,parent,l_child,r_child,mode='entropy'):
        l_weight=len(l_child)/len(parent)
        r_weight=len(r_child)/len(parent)
        
        if mode=='gini':
            gain=self.gini_index(parent)-(l_weight*self.gini_index(l_child)+r_weight*self.gini_index(r_child))
            
        else:
            gain=self.entropy(parent)-(l_weight*self.entropy(l_child)+r_weight*self.entropy(r_child))
        
        return gain
    def gini_index(self,y):
        gini=0
        class_label=np.unique(y)
        
        for cls in class_label:
            p_cls=len(y[y==cls])/len(y)
            gini+=p_cls**2
            
        return 1-gini
    
    def entropy(self,y):
        entropy=0
        class_label=np.unique(y)
        
        for cls in class_label:
            p_cls=len(y[y==cls])/len(y)
            entropy+= -p_cls*np.log2(p_cls)
        
        return entropy
    
    def get_leaf_value(self,y):
        Y=list(y)
        
        return max(Y,key=Y.count)
            
    def fit(self,x,y):
        X=pd.DataFrame(x)
        X.insert(19,'Stage',y)
        data=np.array(X)
        self.root=self.creating_tree(data)
        
    def prediction(self,x):
        prediction=[self.make_prediction(x,self.root) for x in X]
        return prediction
    
    def make_prediction(self,x,tree):
        if tree.value!=None:
            return tree.value
        feature_val=x[tree.feature_index]
        if feature_val<=tree.threshold:
            return self.make_prediction(x.tree.left)
        if feature_val>tree.threshold:
            return self.make_prediction(x.tree.right)
        
            
```
